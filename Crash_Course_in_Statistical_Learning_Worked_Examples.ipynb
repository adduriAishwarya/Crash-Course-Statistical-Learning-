There are null values, so instead of dropping rows, lets perform simple random imputation

**Simple Random Imputation** (we impute the missing data with random observed values of the variable) which is later followed by Regression Imputation of each of the variables iteratively.

**Performing Random Imputation**

```python
def random_imputation(data, feature):

    number_missing = data[feature].isnull().sum()
    observed_values = data.loc[data[feature].notnull(), feature]
    data.loc[data[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing, replace = True)
    
    return data
```

```python
missing_columns = ["rear_camera_mp", "front_camera_mp", "internal_memory", "ram", "battery","weight"]
```

```python
for feature in missing_columns:
    data[feature + '_imp'] = data[feature]
    data = random_imputation(data, feature)
```

**Removing the outliers:**

```python
import numpy as np
from scipy import stats

# Calculate the z-score for each data point in selected columns
selected_cols = ['screen_size', '4g','5g', 'release_year', 'days_used', 
             'normalized_used_price', 'normalized_new_price', 'rear_camera_mp_imp', 
             'front_camera_mp_imp', 'internal_memory_imp',
             'ram_imp', 'battery_imp', 'weight_imp', 'Device_Brand']
z = np.abs(stats.zscore(data[selected_cols]))

# Identify and remove outliers using the z-score method
threshold = 3
data_RO = data[(z < threshold).all(axis=1)]
```

**SCATTER PLOT AFTER REMOVING OUTLIERS**

```python
import matplotlib.pyplot as plt
import pandas as pd



# create a scatter plot for each pair of columns
fig, axes = plt.subplots(nrows=len(data_RO.columns), ncols=len(data_RO.columns), figsize=(30, 30))
for i, column1 in enumerate(data_RO.columns):
    for j, column2 in enumerate(data_RO.columns):
        axes[i,j].scatter(data_RO[column1], data_RO[column2])
        axes[i,j].set_xlabel(column1)
        axes[i,j].set_ylabel(column2)

# show the plot
plt.tight_layout()
plt.show()
```

# **NORMALIZATION**

```python
from sklearn.preprocessing import MinMaxScaler
# Create an instance of MinMaxScaler
scaler = MinMaxScaler()
# Normalize all columns
data_RO_MM = pd.DataFrame(scaler.fit_transform(data_RO), columns=data_RO.columns)
# Print the normalized dataframe
print(data_RO_MM)
```

```python
data_RO_MM.head()
```

```python
from sklearn.model_selection import  train_test_split

X = data_RO_MM[['screen_size', '4g', '5g', 'release_year', 
                                                     
                                                   'normalized_new_price', 'rear_camera_mp_imp', 
                                                   'front_camera_mp_imp', 
                                                   'ram_imp',  
                                                   'weight_imp',  'os_Android', 
                                                   'os_Others', 'os_Windows', 'os_iOS']]

y = data_RO_MM['normalized_used_price']

#Spliting data into Training 76.5%, Validation set 13.5% and Test set 10%

X_t, X_test, y_t, y_test = train_test_split(X, y, test_size=0.1, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_t, y_t, test_size=0.15, random_state=1)
```

```python
from sklearn.metrics import r2_score, mean_squared_error 
from sklearn.linear_model import LinearRegression
from sklearn import datasets, linear_model
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train,y_train)
```

```python
# Make predictions using the training set
y_pred = regr.predict(X_train)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_train, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_train, y_pred))
r2 = r2_score(y_train,y_pred)
print('R^2 score on tarining set =',r2)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))
r2 = r2_score(y_test,y_pred)
print('R^2 score on test set =',r2)
```

```python
# Make predictions using the validation set
y_pred = regr.predict(X_val)

# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_val, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_val, y_pred))
r2_val = r2_score(y_val,y_pred)
print('R^2 score on validation set =',r2_val)
```
